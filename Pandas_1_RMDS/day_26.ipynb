```python
# pandas split functionality
import pandas as pd
df=pd.DataFrame({
    'card':['1971 Topps Nolan Ryan','1928 Ogden Don Bradman','1909 T206 Ty Cobb','1887 Lone Ben Franklin','2005 Topps Justin Verlander']
})
df
```

```python
# generic string split
df['card'].str.split(' ')
```

```python
# expand=True to get all the split data into the dataframe 
df['card'].str.split(' ',expand=True)
```

```python
# string split using N=1
df['card'].str.split(' ',n=1,expand=True)
```

```python
df['card'].str.split(' ',n=2,expand=True)
```

```python
# rename the columns 
df['card'].str.split(' ',expand=True).rename(columns={0:'year',1:'brand',2:'subject',3:'last'})
```

```python
df.head()
```

```python
df[['year','brand','subject']]=df['card'].str.split(' ',n=2,expand=True)
```

```python
df
```

```python
df=df.drop(columns=['card'])
```
