```python
# Rolling Average
df['rolling_mean_3'] = df['Capacity'].rolling(window=3).mean()
df.head(10)
```

```python
# Rolling Sum
df['rolling_sum'] = df['Capacity'].rolling(window=3).sum()
df.head(10)
```

```python
# Rolling Agg
rolling_stats = df['Capacity'].rolling(window=5).agg(['mean', 'sum', 'std'])
rolling_stats.head(10)
```

```python
# Apply
# df['rolling_range'] = df['tickets_sold'].rolling(window=3).apply(lambda x: x.max() - x.min())

# df['Rolling_mean_venue'] = df.groupby('Venue')['tickets_sold'].rolling(window=2).mean().reset_index(level=0, drop=True)

df['rolling_avg_min_periods'] = df['Capacity'].rolling(window=5, min_periods=3).mean()
df.head(10)
```

```python
df['rolling_shift'] = df['Capacity'].rolling(window=5).mean().shift(1) - df['Capacity']
df.head(10)
```

```python
df['rolling_filled_values'] = df['Capacity'].ffill().rolling(window=5).mean()
df.head(10)
```

```python
df['centered_rolling_avg'] = df['Capacity'].rolling(window=5, center=True).mean()
df['centered_rolling_avg'].head(10)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df['rolling_avg_7_days'] = df['Capacity'].rolling('7D').mean()
```
