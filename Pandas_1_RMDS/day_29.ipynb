```python
# pandas explode
# Using explode within Python Pnadas allows you to transform each element of a list to a new row within a dataframe.

import pandas as pd
df = pd.DataFrame({  'card': ['1971 Nolan Ryan', '1928 Ogden Don Bradman', '1909 T206 Ty Cobb', '1887 Lone Jack Ben Franklin', '2005 Topps Justin Verlander'],
'properties': [['Baseball', 'Vintage', 'Pitcher'], ['Cricket', 'Pre War'], ['Baseball', 'Pre War', 'Batter'], ['Non Sports', 'Pre War'], ['Baseball', 'Modern', 'Pitcher']]
              })

df
```

```python
df = df.explode('properties')
df
```

```python
df['properties'].value_counts()['Pre War']
```

```python
df.query('properties == "Baseball"')
```

```python
df.pivot_table(index='properties', values='card', aggfunc='count').reset_index().rename(columns = {'card':'count'}).sort_values('count', ascending = False)
```
