```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# python pandas group by
df = sns.load_dataset('tips')
df.head(5)
```

```python
# Groupby Single Calculations
df.groupby('time')['total_bill'].mean()
```

```python
# If you want to see the above represented nicer in a table use to_frame()
df.groupby('time')['total_bill'].mean().to_frame('Avg_bill')
```

```python
# When creating a new dataframe from the groupby youâ€™ll often see reset_index used. 
df.groupby('time')['total_bill'].mean().to_frame('Avg_bill').reset_index()
```

```python
df.groupby('time')['total_bill'].mean().to_frame('Avg_bill').reset_index().sort_values('Avg_bill', ascending = False)
```

```python
# Count: Number of non-null entries in each column or row
df.groupby('time')['total_bill'].count()
```

```python
# Size: Number of entries in each column or row
df.groupby('time')['total_bill'].size()
```

```python
df.groupby('time')['total_bill'].nunique()
```

```python
# uestion 1: Find the Average Tip for each night of the week
df.groupby('day')['tip'].mean()
```

```python
# Question 2: Find the Max Tip for each sex
df.groupby('sex')['tip'].max()
```

```python
# Question 3: What's considered a top 10% tip for each night
df.groupby('day')['tip'].quantile(.9)
```

```python
# Multiple Calculations
df.groupby('time')['total_bill'].agg(['min', 'max'])
```

```python
df.groupby('time')['total_bill'].agg([min, 'max', np.median])
```

```python
dftransform = sns.load_dataset('tips')
```

```python
dftransform['mediumtips'] = dftransform.groupby('time')['total_bill'].transform('median')
```

```python
dftransform
```

```python
dftransform[dftransform['mediumtips'] < dftransform['total_bill']]
```

```python
dftransform['zscore'] = dftransform.groupby('time')['total_bill'].transform(lambda x: (x - x.mean()) / x.std())
```

```python
dftransform
```

```python
dftransform.groupby('time').apply(lambda x: (x['tip']/x['total_bill']).mean()).to_frame('tip_percent').reset_index()
```

```python
tx = sns.load_dataset('taxis')
tx.head(10)
```

```python
tx.groupby(tx['pickup'].dt.month)['distance'].mean()
```

```python
# Q4 Find the Interquartile Range Q3-Q1 for Tips within Boroughs
tx.groupby('pickup_borough').apply(lambda x: x['tip'].quantile(.75) - x['tip'].quantile(.25))
```

```python
# Q5 Find the Mean, Min, Max Fare for Payment Type and Taxi Color
tx.groupby(['payment', 'color'])['fare'].agg(avg = 'mean', minimum='min', maximum='max')
```

```python
# Q6a Add the average total for a taxi color to each row in the dataframe
tx['avgtotal'] = tx.groupby('color')['total'].transform('mean')
```

```python
# Q6b Only Show the rows that have a greater total than the average
tx.query('total > avgtotal')
```
